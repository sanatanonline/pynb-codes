```python
movies = data.movies()
movies.head()
```

```python
movies.describe()
```

```python
movies.Director.value_counts()
```

```python
directors = movies.groupby(['Creative_Type', 'Director']).agg('Production_Budget').sum().reset_index()
directors.head()
```

```python
directors = directors.reset_index()
directors.head()
```

```python
dirs_pivoted = directors.pivot_table(
    index='Director', columns='Creative_Type', values='Production_Budget', aggfunc='sum')
dirs_pivoted.head()
```
