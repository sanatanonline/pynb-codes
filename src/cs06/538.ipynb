```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
edu_data = pd.read_csv('http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv')
edu_data.head()
```

```python
under_20_pct = edu_data.loc[0, edu_data.iloc[0] < 20].index
under_20_pct
```

```python
edu_data.plot(x='Year', y='Agriculture', figsize=(15, 10))
```

```python
style.use('fivethirtyeight')
edu_plot = edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
```

```python
edu_plot = edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
edu_plot.tick_params(axis='both', which='major', labelsize=18)
```

```python
edu_plot = edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
edu_plot.tick_params(axis='both', which='major', labelsize=18)
edu_plot.set_yticklabels(labels = ['santa', '0 ', '10 ', '20 ', '30 ', '40 ', '50% ', '60'])
plt.show()
```

```python
edu_plot = edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
edu_plot.tick_params(axis='both', which='major', labelsize=18)
edu_plot.set_yticklabels(labels = ['santa', '0 ', '10 ', '20 ', '30 ', '40 ', '50% ', '60'])
edu_plot.set_xlim(left=1969, right=2011)
plt.show()
```

```python
'''colorblind-friendly colors'''
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [213/255,94/255,0], [0,114/255,178/255]]

edu_plot = edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10), legend=False)
edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
edu_plot.tick_params(axis='both', which='major', labelsize=18)
edu_plot.set_yticklabels(labels = ['santa', '0 ', '10 ', '20 ', '30 ', '40 ', '50% ', '60'])
edu_plot.set_xlim(left=1969, right=2011)

edu_plot.text(x=1966,
              y=-10,
              s="""© Sanatan Das                                                                           """
                + """                                                        """
                + """Source: Lending Club""",
              fontsize=17,
              color='#f0f0f0',
              backgroundcolor='grey'
              )


# Adding a title and a subtitle
edu_plot.text(x = 1966.65, y = 62.7, s = "The gender gap is transitory - even for extreme cases",
              fontsize = 26, weight = 'bold', alpha = .75)
edu_plot.text(x = 1966.65, y = 57,
              s = 'Percentage of Bachelors conferred to women from 1970 to 2011 in the US for\nextreme cases where the percentage was less than 20% in 1970',
              fontsize = 19, alpha = .85)


# Add colored labels
edu_plot.text(x = 1994, y = 43, s = 'Agriculture', color = colors[0], weight = 'bold', rotation = 33,
              backgroundcolor = '#f0f0f0')
edu_plot.text(x = 1985, y = 42.2, s = 'Architecture', color = colors[1], weight = 'bold', rotation = 18,
              backgroundcolor = '#f0f0f0')
edu_plot.text(x = 2004, y = 51, s = 'Business', color = colors[2], weight = 'bold', rotation = -5,
               backgroundcolor = '#f0f0f0')
edu_plot.text(x = 2001, y = 30, s = 'Computer Science', color = colors[3], weight = 'bold', rotation = -42.5,
              backgroundcolor = '#f0f0f0')
edu_plot.text(x = 1987, y = 11.5, s = 'Engineering', color = colors[4], weight = 'bold',
              backgroundcolor = '#f0f0f0')
edu_plot.text(x = 1976, y = 23, s = 'Physical Sciences', color = colors[5], weight = 'bold', rotation = 27,
              backgroundcolor = '#f0f0f0')

plt.show()
```
